# Replacing the null values

```python
arr
```

```python
# Replace all the nan values by 0
```

```python
arr[np.isnan(arr)]=0
```

```python
arr
```

```python
# Replace all the nan values by the mean value
```

```python
arr=np.array([[10., 20., 30., np.nan],
       [50., np.nan, 45., 60.],
       [np.nan, np.nan, 62., 78.]])
```

```python
arr
```

```python
arr.mean()
```

```python
m=arr[~(np.isnan(arr))].mean()
m
```

```python
arr[np.isnan(arr)]=m
```

```python
arr
```

# np.argmax()

- It returns the index of the maximum number.
- In case of multiple maximum. It will return the index of the first maximum number.
- It treates the nd array as 1-D array

```python
arr=np.array([10,-20,90,-50,0,70,90,0,82]).reshape(3,3)
arr
```

```python
arr.max()
```

```python
np.argmax(arr)
```

```python
arr.ravel()
```

```python
arr=np.array([10,-20,40,-50,0,70,90,0,82]).reshape(3,3)
arr
```

```python
np.argmax(arr)
```

```python
arr.ravel()
```
