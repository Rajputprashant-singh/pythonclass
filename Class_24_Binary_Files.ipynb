# Binary Files
- A binary file stores the data in bytes format.
- The data has to be converted into bytes then only we can store it in a binary file.

# modes

```python
# read mode - 'rb'
# write - 'wb'
# append mode - 'ab'
# read and write mode - 'rb+'
# write and read mode - 'wb+'
# append and read mode - 'ab+'
```

```python
lst=[10,20,30,40,50]

# Store the above list in a binary file lst.dat

with open('lst.dat','wb+') as f:
    f.write(lst)
```

# bytes() and bytearray()

# bytes()

- It converts the integer values into bytes format.
- It can convert the int from 0 to 256 to bytes format.
- It is immutable.

# bytearray()

- It converts the int values into bytes format.
- It can only convert the int from 0 to 256.
- It is mutable.

- We have seen that bytes() and bytearray() have a lot of limitations.
- To overcome the limitations of bytes() and bytearray() we use pickling and unpickling method.

# Pickling and Unpickling

# Pickling
- It is also known as serialization. It is the process of converting any type of data into the bytes format and storing it in a binary file.

# Unpickling
- It is also known as deserialization. It is the process of reading the data from the file and converting it into its actual format.

- We use pickle module to pickle or unpickle the data

```python
import pickle
```

# dump()

- It is used to pickle the data.

```python
# pickle.dump(data,file_object)
```

```python
lst=[500,600,1000,2000]

# Store the above list in lst.dat file
```

```python
with open('lst.dat','wb') as f:
    pickle.dump(lst,f)
```

```python
# ls
```

```python
with open('lst.dat','rb') as f:
    x=f.read()
```

```python
x
```

# load()

- It is used to unpickle the data.

```python
# pickle.load(file_object)
```

```python
with open('lst.dat','rb') as f:
    print(pickle.load(f))
```

```python
emp={'name':'Aniket','Age':26,'City':'Bangalore','Salary':95000}
emp
```

```python
with open('emp.dat','wb+') as f:
    pickle.dump(emp,f)
    f.seek(0)
    print(f.read())
```

```python
with open('emp.dat','rb+') as f:
    print(pickle.load(f))
```

# How for loop works

- First it coneverts the iterable into an iterator.
- Second it calls the next() again and again on the iterator object until the length of the iterator is exhausted.

```python
my_str='Python'

for i in my_str:
    print(i)
```
