# Indexing and slicing of arrays

# Indexing and slicing of 1-D array

- The indexing and slicing of a 1-D array is similar to the indexing and slicing of a list.
- Array supports positive and negative indexing both.

# Indexing

```python
# arr[index]
```

```python
arr=np.array([10,20,30,40,50,60])
arr
```

```python
arr.shape
```

```python
arr.ndim
```

```python
# Fetch the first item from the array

print(arr[0])
print(arr[-6])
print(arr[-len(arr)])
```

```python
-len(arr)
```

```python
arr
```

```python
# Fetch the last element from the array

print(arr[-1])
print(arr[5])
print(arr[len(arr)-1])
```

```python
len(arr)-1
```

# Slicing

```python
# arr[start:stop:step]
# Stop index is excluded
```

```python
arr
```

```python
# Fetch the first four elements.

print(arr[0:4:1])
print(arr[0:4])
print(arr[:4])
```

```python
arr
```

```python
# Reverse the array

print(arr[::-1])
```

```python
# Every second element in the reverse order
arr[::-2]
```
